#Preprocessing of Data

```python
#target
y=data['price']
```

```python
#Messing Data
data.isna().mean()
```

```python
numerical_data=['wheelbase','carlength','carwidth','curbweight','enginesize','boreratio','stroke','horsepower','citympg','highwaympg']
categorical_data=['carbody','cylindernumber']
bin_data=['fueltype','aspiration','doornumber','enginelocation']
all_indx=['wheelbase','carlength','carwidth','curbweight','enginesize','boreratio','stroke','horsepower','citympg','highwaympg','carbody','cylindernumber','fueltype','aspiration','doornumber','enginelocation']
```

```python
#Data Encoding (LabelEncoder)
from sklearn.preprocessing import LabelEncoder
label_encod=LabelEncoder()
data[bin_data[0]]=label_encod.fit_transform(data[bin_data[0]])
data[bin_data[1]]=label_encod.fit_transform(data[bin_data[1]])
data[bin_data[2]]=label_encod.fit_transform(data[bin_data[2]])
data[bin_data[3]]=label_encod.fit_transform(data[bin_data[3]])
```

```python
data_all=data[all_indx]
```

```python
#Data Encoding (OnHotEncoder)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
colTrans=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),categorical_data)],remainder="passthrough") 
data_all=colTrans.fit_transform(data_all)
```

```python
#Spliting of data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data_all,y,train_size=0.7)
```

```python
x_test
```

#**Train Model**

```python
from sklearn.linear_model import LinearRegression
linear_reg=LinearRegression()
linear_reg.fit(x_train,y_train)
```

```python
y_pred=linear_reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print(f"Score = {r2_score(y_test,y_pred)*100}%")
```
